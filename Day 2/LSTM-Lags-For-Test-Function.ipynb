```python
from datetime import date
def week_cal(x):
    if date(2021, 10, 4) <= x <= date(2021, 10, 10):
        return "w01"
    elif date(2021, 10, 11) <= x <= date(2021, 10, 17):
        return "w02"
    elif date(2021, 10, 18) <= x <= date(2021, 10, 24):
        return "w03"
    elif date(2021, 10, 25) <= x <= date(2021, 10, 31):
        return "w04"
    elif date(2021, 11, 1) <= x <= date(2021, 11, 7):
        return "w05"
    elif date(2021, 11, 8) <= x <= date(2021, 11, 14):
        return "w06"
    elif date(2021, 11, 15) <= x <= date(2021, 11, 21):
        return "w07"
    elif date(2021, 11, 22) <= x <= date(2021, 11, 28):
        return "w08"
    elif date(2021, 11, 29) <= x <= date(2021, 12, 5):
        return "w09"
    elif date(2021, 12, 6) <= x <= date(2021, 12, 12):
        return "w10"
    elif date(2021, 12, 13) <= x <= date(2021, 12, 19):
        return "w11"
    elif date(2021, 12, 20) <= x <= date(2021, 12, 26):
        return "w12"
    elif date(2021, 12, 27) <= x <= date(2022, 1, 2):
        return "w13"
    elif date(2022, 1, 3) <= x <= date(2021, 1, 9):
        return "w14"
    elif date(2022, 1, 10) <= x <= date(2022, 1, 16):
        return "w15"
    elif date(2022, 1, 17) <= x <= date(2022, 1, 23):
        return "w16"
    elif date(2022, 1, 24) <= x <= date(2022, 1, 30):
        return "w17"
    elif date(2022, 1, 31) <= x <= date(2022, 2, 6):
        return "w18"
    elif date(2022, 2, 7) <= x <= date(2022, 2, 13):
        return "w19"
    else:                                                       # 28 <= x or x <= 6
        return 'Omit'
    # Add elif statements....

train['Week'] = train['DateID'].apply(week_cal)
```

```python
from keras.preprocessing import sequence
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout, Flatten, AveragePooling1D

from keras.optimizers import Adam
from keras.models import load_model
from keras.callbacks import ModelCheckpoint

def get_model(shape1, shape2):
    model = Sequential()
    model.add(LSTM(6, batch_input_shape=(1, shape1, shape2), stateful=True))
    model.add(Dense(1))
    model.compile(loss='mean_absolute_percentage_error', optimizer='adam')
    return model
```

```python
df_prediction = pd.DataFrame(columns=['CategoryCode', 'ItemCode', 'Week', 'PredictedSales'])
i = 0
def week_cal(df, item_code, cat_code):
    print('week cal running')
    global i
    global df_prediction
    CategoryCode = cat_code
    ItemCode = item_code
    w1 = df['pred_value'].iloc[0]
    w2 = df['pred_value'].iloc[1]
    w3 = df['pred_value'].iloc[2]
    w4 = df['pred_value'].iloc[3]
#     print('CategoryCode',CategoryCode, 'ItemCode',ItemCode, 'Week','w1', 'PredictedSales',w1)
    df_prediction.loc[i] = [CategoryCode] +  [ItemCode] + ['w1'] + [w1]
    i = i+1
    df_prediction.loc[i] = [CategoryCode] +  [ItemCode] + ['w2'] + [w2]
    i = i+1
    df_prediction.loc[i] = [CategoryCode] +  [ItemCode] + ['w3'] + [w3]
    i = i+1
    df_prediction.loc[i] = [CategoryCode] +  [ItemCode] + ['w4'] + [w4]
    i = i+1
#     print(df_prediction)
```

```python
#import MinMaxScaler and create a new dataframe for LSTM model
from sklearn.preprocessing import MinMaxScaler

def predict(item, item_code, cat_code):
    item = item.drop(columns=['CategoryCode','ItemCode'])
    item = item.sort_values(by='DateID', ascending=True).reset_index(drop=True)
    item = item.fillna(item['WeeklySales'].min())
    print(item)
    #create a new dataframe to model the difference
    df_diff = item.copy()
    #add previous sales to the next row
    df_diff['prev_sales'] = df_diff['WeeklySales'].shift(1)
    #drop the null values and calculate the difference
    df_diff = df_diff.dropna()
    df_diff['diff'] = (df_diff['WeeklySales'] - df_diff['prev_sales'])
    df_diff
    
    #create dataframe for transformation from time series to supervised
    df_supervised = df_diff.drop(['prev_sales'],axis=1)
    #adding lags
    for inc in range(1,10):
        field_name = 'lag_' + str(inc)
        df_supervised[field_name] = df_supervised['diff'].shift(inc)
    #drop null values
    df_supervised = df_supervised.dropna().reset_index(drop=True)
    
    #MinMaxScaler and create a new dataframe for LSTM model
    df_model = df_supervised.drop(['DateID'],axis=1)
    #split train and test set
    train_set, test_set = df_model.iloc[0:-4].values, df_model.iloc[-4:].values
    
    #apply Min Max Scaler
    scaler = MinMaxScaler(feature_range=(-1, 1))
    scaler = scaler.fit(train_set)
    # reshape training set
    train_set = train_set.reshape(train_set.shape[0], train_set.shape[1])
    train_set_scaled = scaler.transform(train_set)
    # reshape test set
    test_set = test_set.reshape(test_set.shape[0], test_set.shape[1])
    test_set_scaled = scaler.transform(test_set)
    
    X_train, y_train = train_set_scaled[:, 1:], train_set_scaled[:, 0:1]
    X_train = X_train.reshape(X_train.shape[0], 1, X_train.shape[1])
    X_test, y_test = test_set_scaled[:, 1:], test_set_scaled[:, 0:1]
    X_test = X_test.reshape(X_test.shape[0], 1, X_test.shape[1])
#     print(y_test)
    #get model
    model = get_model(X_train.shape[1], X_train.shape[2])
    model.fit(X_train, y_train, nb_epoch=100, batch_size=1, verbose=0, shuffle=False)
    
    y_pred = model.predict(X_test,batch_size=1)
    #for multistep prediction, you need to replace X_test values with the predictions coming from t-1
    
    #reshape y_pred
    y_pred = y_pred.reshape(y_pred.shape[0], 1, y_pred.shape[1])
    #rebuild test set for inverse transform
    pred_test_set = []
    for index in range(0,len(y_pred)):
#         print (np.concatenate([y_pred[index],X_test[index]],axis=1))
        pred_test_set.append(np.concatenate([y_pred[index],X_test[index]],axis=1))
    #reshape pred_test_set
    pred_test_set = np.array(pred_test_set)
    pred_test_set = pred_test_set.reshape(pred_test_set.shape[0], pred_test_set.shape[2])
    #inverse transform
    pred_test_set_inverted = scaler.inverse_transform(pred_test_set)
    
    #create dataframe that shows the predicted sales
    result_list = []
    sales_dates = list(item[-5:].DateID)
    act_sales = list(item[-5:].WeeklySales)
    for index in range(0,len(pred_test_set_inverted)):
        result_dict = {}
        result_dict['pred_value'] = int(pred_test_set_inverted[index][0] + act_sales[index])
        result_dict['DateID'] = sales_dates[index+1]
        result_list.append(result_dict)
    df_result = pd.DataFrame(result_list)
    #for multistep prediction, replace act_sales with the predicted sales
#     print(df_result)
    week_cal(df_result, item_code, cat_code) 
```
