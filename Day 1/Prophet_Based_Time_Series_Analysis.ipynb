```python
df_prediction = pd.DataFrame(columns=['CategoryCode', 'ItemCode', 'Week', 'PredictedSales'])
df_prediction
```

```python
i = 0
def week_cal(df, item_code, cat_code):
    print('week cal running')
    global i
    global df_prediction
    CategoryCode = cat_code
    ItemCode = item_code
    w1 = df['yhat'].iloc[0:7].sum()
    w2 = df['yhat'].iloc[7:14].sum()
    w3 = df['yhat'].iloc[14:21].sum()
    w4 = df['yhat'].iloc[21:].sum()
    print('CategoryCode',CategoryCode, 'ItemCode',ItemCode, 'Week','w1', 'PredictedSales',w1)
    df_prediction.loc[i] = [CategoryCode] +  [ItemCode] + ['w1'] + [w1]
    i = i+1
    df_prediction.loc[i] = [CategoryCode] +  [ItemCode] + ['w2'] + [w2]
    i = i+1
    df_prediction.loc[i] = [CategoryCode] +  [ItemCode] + ['w3'] + [w3]
    i = i+1
    df_prediction.loc[i] = [CategoryCode] +  [ItemCode] + ['w4'] + [w4]
    i = i+1
```

```python
import pandas as pd
from fbprophet import Prophet
def forecast(df):
#   df = pd.read_csv(data_frame_path)
  df.rename(columns = {'DateID' : 'ds','DailySales' : 'y'}, inplace = True)
  model = Prophet(interval_width=0.7)
  model.fit(df)
  future = model.make_future_dataframe(periods=28)
  forecast = model.predict(future )
  p=forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
  return p
```
